### Demostración de métodos de lista

```python
# 1. append(): Añade un elemento al final de la lista.
mi_lista_demo = [1, 2, 3]
print(f"Lista original: {mi_lista_demo}")
mi_lista_demo.append(4)
print(f"Después de append(4): {mi_lista_demo}")
```

```python
# 2. insert(): Inserta un elemento en una posición específica.
mi_lista_demo = [1, 2, 3, 4]
print(f"Lista original: {mi_lista_demo}")
mi_lista_demo.insert(1, 99)
print(f"Después de insert(1, 99): {mi_lista_demo}")
```

```python
# 3. extend(): Añade los elementos de otra lista (o cualquier iterable) al final de la lista actual.
mi_lista_demo = [1, 2, 3]
otra_lista = [5, 6]
print(f"Lista original: {mi_lista_demo}")
print(f"Otra lista: {otra_lista}")
mi_lista_demo.extend(otra_lista)
print(f"Después de extend(otra_lista): {mi_lista_demo}")
```

```python
# 4. remove(): Elimina la primera aparición de un valor específico.
mi_lista_demo = [1, 2, 3, 2, 4]
print(f"Lista original: {mi_lista_demo}")
mi_lista_demo.remove(2)
print(f"Después de remove(2): {mi_lista_demo}")
```

```python
# 5. pop(): Elimina y devuelve el elemento en la posición dada. Si no se especifica índice, elimina y devuelve el último elemento.
mi_lista_demo = [1, 2, 3, 4]
print(f"Lista original: {mi_lista_demo}")
elemento_eliminado = mi_lista_demo.pop(1)
print(f"Elemento eliminado con pop(1): {elemento_eliminado}")
print(f"Lista después de pop(1): {mi_lista_demo}")
elemento_eliminado_final = mi_lista_demo.pop()
print(f"Elemento eliminado con pop(): {elemento_eliminado_final}")
print(f"Lista después de pop(): {mi_lista_demo}")
```

```python
# 6. count(): Devuelve el número de veces que un valor aparece en la lista.
mi_lista_demo = [1, 2, 2, 3, 2, 4]
print(f"Lista original: {mi_lista_demo}")
conteo_dos = mi_lista_demo.count(2)
print(f"El número 2 aparece {conteo_dos} veces.")
```

```python
# 7. index(): Devuelve el índice de la primera aparición de un valor.
mi_lista_demo = [1, 2, 3, 2, 4]
print(f"Lista original: {mi_lista_demo}")
indice_dos = mi_lista_demo.index(2)
print(f"El primer 2 está en el índice: {indice_dos}")
```

```python
# 8. sort(): Ordena los elementos de la lista en su lugar (modifica la lista original).
mi_lista_demo = [3, 1, 4, 1, 5, 9, 2]
print(f"Lista original: {mi_lista_demo}")
mi_lista_demo.sort()
print(f"Después de sort(): {mi_lista_demo}")
mi_lista_demo.sort(reverse=True)
print(f"Después de sort(reverse=True): {mi_lista_demo}")
```

```python
# 9. reverse(): Invierte el orden de los elementos de la lista en su lugar.
mi_lista_demo = [1, 10, 8, 4, 5]
print(f"Lista original: {mi_lista_demo}")
mi_lista_demo.reverse()
print(f"Después de reverse(): {mi_lista_demo}")
```

```python
# 10. copy(): Devuelve una copia superficial de la lista.
mi_lista_original = [1, 2, 3]
mi_lista_copia = mi_lista_original.copy()
print(f"Lista original: {mi_lista_original}")
print(f"Copia de la lista: {mi_lista_copia}")
# Demostrar que son objetos diferentes
mi_lista_original.append(4)
print(f"Lista original después de append: {mi_lista_original}")
print(f"Copia de la lista (no afectada): {mi_lista_copia}")
```

```python
# 11. clear(): Elimina todos los elementos de la lista.
mi_lista_demo = [1, 2, 3, 4]
print(f"Lista original: {mi_lista_demo}")
mi_lista_demo.clear()
print(f"Después de clear(): {mi_lista_demo}")
```
